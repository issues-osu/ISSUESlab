```python
import pandas as pd
import geopandas as gpd
import shapely.geometry
import json

df = pd.read_csv("pids_urls.csv")

def geojson_to_geom(geojson_str):
    try:
        fixed_str = geojson_str.replace("'", '"')
        geojson_obj = json.loads(fixed_str)
        return shapely.geometry.shape(geojson_obj)
    except Exception as e:
        print(f"Error parsing geometry: {e}")
        return None

df['geometry'] = df['geometry'].apply(geojson_to_geom)

gdf = gpd.GeoDataFrame(df, geometry='geometry')

gdf['lat'] = gdf.geometry.y
gdf['lon'] = gdf.geometry.x

gdf = gdf.rename(columns={'id': 'panoid'})

pano_df = gdf[['panoid', 'lat', 'lon']]

pano_df.to_csv("c:/users/barboza-salerno.1/Downloads/neglect/cleaned_pids.csv", index=False)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import shape
import matplotlib.pyplot as plt
import json

# Load your CSV
df = pd.read_csv(r"c:/users/barboza-salerno.1/Downloads/neglect/pids_urls.csv")

# Fix geometry column by replacing single quotes with double quotes
def fix_geometry_string(s):
    try:
        # Replace single quotes with double quotes
        s_fixed = s.replace("'", '"')
        # Parse JSON string to dict
        return json.loads(s_fixed)
    except Exception as e:
        print(f"Error parsing geometry: {e}")
        return None

# Apply fix
df['geometry_fixed'] = df['geometry'].apply(fix_geometry_string)

# Drop rows with failed geometry parsing
df = df.dropna(subset=['geometry_fixed'])

# Convert dict geometry to shapely geometry
df['geometry'] = df['geometry_fixed'].apply(shape)

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry')

# Extract lat/lon
gdf['lat'] = gdf.geometry.y
gdf['lon'] = gdf.geometry.x

# Rename id to panoid for plotting usage
gdf = gdf.rename(columns={'id': 'panoid'})

# Save cleaned panoid CSV with lat/lon if needed
gdf[['panoid', 'lat', 'lon']].to_csv(r"c:/users/barboza-salerno.1/Downloads/neglect/mly_pids_fixed.csv", index=False)

# Quick plot to check points
gdf.plot(marker='o', color='red', markersize=5)
plt.title("Mapillary Points")
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Assuming you have your GeoDataFrame 'gdf' with geometry column

# Assign CRS to WGS84 lat/lon if not set
gdf = gdf.set_crs(epsg=4326, allow_override=True)

# Project to Web Mercator for metric units
gdf_proj = gdf.to_crs(epsg=3857)

# Extract x and y coords
x = gdf_proj.geometry.x
y = gdf_proj.geometry.y

# Plot hexbin
fig, ax = plt.subplots(figsize=(10,10))
hb = ax.hexbin(x, y, gridsize=50, cmap='viridis', mincnt=1)
cb = fig.colorbar(hb, ax=ax)
cb.set_label('Number of Points')
ax.set_axis_off()
plt.title('Hexbin Map of Points')
plt.show()
```

```python
import pandas as pd
import json

# Load pids_urls.csv
df = pd.read_csv("c:/users/barboza-salerno.1/downloads/neglect/pids_urls.csv")

# Function to parse GeoJSON-like geometry string and extract lat/lon
def extract_lat_lon(geom_str):
    try:
        # Replace single quotes with double quotes if needed
        geom = json.loads(geom_str.replace("'", '"'))
        # GeoJSON Point: coordinates [lon, lat]
        lon, lat = geom['coordinates']
        return lat, lon
    except Exception as e:
        print(f"Error parsing geometry: {e}")
        return None, None

df['lat'], df['lon'] = zip(*df['geometry'].apply(extract_lat_lon))

# Drop rows where parsing failed
df = df.dropna(subset=['lat', 'lon'])

# Keep only needed columns for plot_map
pid_df = df[['id', 'lat', 'lon']]

# Save cleaned pano ID file with lat/lon
pid_df.to_csv("c:/users/barboza-salerno.1/downloads/neglect/mly_pano_id_latlon.csv", index=False)
```
